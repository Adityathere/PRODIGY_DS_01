```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker
```

```python
df = pd.read_csv(r"C:\Users\ADITYA\Desktop\Datasets\world_population.csv")
df.head(10)
```

```python
columns_of_interest = ['Continent', '2022 Population', '1970 Population']
data = df[columns_of_interest].dropna()
data['2022 Population'] = pd.to_numeric(data['2022 Population'], errors='coerce')
data['1970 Population'] = pd.to_numeric(data['1970 Population'], errors='coerce')
data['Growth Rate'] = ((data['2022 Population'] - data['1970 Population']) / data['1970 Population'])
aggregated_data = data.groupby('Continent')['Growth Rate'].mean().reset_index()
display(aggregated_data[['Continent', 'Growth Rate']])
```

```python
plt.figure(figsize=(10, 6))
sns.set_style("white")
bars=plt.bar(data['Continent'], data['Growth Rate'], color='orange')
plt.title('Population Growth Rate by Continent')
plt.xlabel('Continent')
plt.ylabel('Growth Rate')
plt.ylim(bottom=0)
for bar in bars:
    bar.set_edgecolor('none')
plt.show()
```

```python
df['2022 Population (Millions)'] = df['2022 Population'] / 1e6
sns.set_style("white")
top_countries = df.sort_values(by='2022 Population', ascending=False).head(20)
plt.figure(figsize=(10, 6))
plt.bar(top_countries['Country/Territory'], top_countries['2022 Population (Millions)'], color='orange')
plt.title('Countries with the Highest Population')
plt.xlabel('Country/Territory')
plt.ylabel('2022 Population (Millions)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
largest_countries = df.sort_values(by='Area (km²)', ascending=False).head(10)
sns.set_style("white")
plt.figure(figsize=(10, 6))
plt.bar(largest_countries['Country/Territory'], largest_countries['Area (km²)'], color='darkcyan')
plt.title('Countries with the Largest Land Area')
plt.xlabel('Country')
plt.ylabel('Land Area (sq km)')
plt.xticks(rotation=45, ha='right')
plt.gca().yaxis.set_major_formatter(lambda x, _: '{:.0f} sq km'.format(x))
plt.show()
```

```python
columns_of_interest = ['Country/Territory', '1970 Population', '1980 Population', '1990 Population', '2000 Population',
                       '2010 Population', '2015 Population', '2020 Population', '2022 Population']
selected_countries = ['China', 'India', 'United States', 'Brazil', 'Russia'
                      , 'Australia', 'Italy', 'Japan', 'Germany', 'Indonesia', 'Nigeria']
df_selected = df[df['Country/Territory'].isin(selected_countries)][columns_of_interest]
df_melted = pd.melt(df_selected, id_vars='Country/Territory', var_name='Year', value_name='Population')
df_melted['Population'] = pd.to_numeric(df_melted['Population'], errors='coerce')

plt.figure(figsize=(12, 6))
palette = sns.color_palette("tab10", n_colors=len(selected_countries))
sns.set_style("darkgrid")

for country, color in zip(selected_countries, palette):
    country_data = df_melted[df_melted['Country/Territory'] == country]
    sns.lineplot(x='Year', y='Population', data=country_data, label=country, color=color)
plt.title('Population Trends Over the Years', fontsize=16)
plt.xlabel('Years', fontsize=14)
plt.ylabel('Population (Millions)', fontsize=14)
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: '{:.0f}M'.format(x / 1e6)))
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```

```python
df_sorted = df.sort_values(by='Density (per km²)', ascending=False)
top_10_capitals = df_sorted.head(10).copy()  
top_10_capitals['Location'] = top_10_capitals['Capital'] +'\n'+top_10_capitals['Country/Territory'] 
plt.figure(figsize=(12, 6))
sns.barplot(x='Location', y='Density (per km²)', hue='Capital', data=top_10_capitals, palette='rocket', legend=False)
plt.title('Top 10 Densely Populated Capitals')
plt.xlabel('Location')
plt.ylabel('Population Density (per km²)')
plt.xticks(rotation=30, ha='right')
plt.show()
```
